```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def height(x,y):
    return(1-x/2 + x**5 + y**3)*np.exp(-x**2 - y**2)
x=np.linspace(-2,2,300)
y=np.linspace(-2,2,300)
X,Y=np.meshgrid(x,y)
plt.contourf(X,Y,height(X,Y),8,alpha=0.8,cmap=plt.cm.hot)
C=plt.contour(X,Y,height(X,Y),8,colors="b")
plt.clabel(C,inline=True,fontsize=8)
```
