```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.array([42829,2749837])
y = np.array([247,329])
X,Y = np.meshgrid(x,y)
Z = np.random.randint(0,8,(2,2))
```

```python
plt.contour(X,Y,Z)
plt.scatter(X.ravel(),Y.ravel(),c=Z.ravel(),cmap="autumn")
```

```python
plt.contourf(X,Y,Z)
plt.scatter(X.ravel(),Y.ravel(),c=Z.ravel(),cmap="autumn")
```
