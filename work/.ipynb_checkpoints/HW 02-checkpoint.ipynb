```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=np.arange(-20,20,0.02)
y=np.arange(-20,20,0.02)
X,Y=np.meshgrid(x,y)
Z=X**2 + Y**2
plt.figure(figsize=(20,12))
contour=plt.contour(X,Y,Z,[20,40],colors="r")
plt.clabel(contour,fontsize=15,colors=("r","b"))
```
